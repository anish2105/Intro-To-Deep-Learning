```python
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np

img = io.imread('zoo2.jpg')
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
red = img[: , : , 0]
green = img[: , : , 1]
blue = img[: , : , 2]
```

```python
plt.imshow(red,cmap = 'Reds')
plt.show()
```

```python
plt.imshow(green,cmap = 'Greens')
plt.show()
```

```python
plt.imshow(blue , cmap = 'Blues')
plt.show()
```

```python
import tensorflow as tf
transformation = tf.keras.preprocessing.image.apply_affine_transform(img , theta = 45 )
plt.imshow(transformation)
plt.show()
```

```python
#Zooming in and zooming out
import tensorflow as tf
transformation = tf.keras.preprocessing.image.apply_affine_transform(img , zx = 0.5 , zy =0.5)
plt.imshow(transformation)
plt.show()
```

```python
#Zooming in and zooming out
import tensorflow as tf
transformation = tf.keras.preprocessing.image.apply_affine_transform(img , tx = 50 , ty =5)
plt.imshow(transformation)
plt.show()
```
